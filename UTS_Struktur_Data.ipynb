Data dari Peminjaman Buku


```python
class NodePinjam:
    def __init__(self, no_antri, id_p, id_b, nama, nim, prodi, judul, tahun, tgl_p, tgl_k):
        self.data = {
            "no_antri": no_antri, # Menyimpan nomor antrean permanen
            "id_pinjam": id_p,
            "id_buku": id_b,
            "nama": nama,
            "nim": nim,
            "prodi": prodi,
            "judul_buku": judul,
            "tahun_terbit": tahun,
            "tanggal_pinjam": tgl_p,
            "tanggal_kembali": tgl_k
        }
        self.next = None
```

Proses dari Antrian

```python
class Linked_List_Perpus:
    # INISIALISASI
    def __init__(self):
        self.head = None #self.head menjelaskan daftar antrian, sedangkan none artinya kosong, jika digabungkan maka daftar antrian sedang kosong
        self.storage = [] # list kosong yang berguna untuk menyimpan data antrian
        self.total_antrean = 0 # Menghitung berapa data yang ada di antrean

    def tambah_data(self, id_p, id_b, nama, nim, prodi, judul, tahun, tgl_p, tgl_k): # Enqueue (menambah data)
        self.total_antrean += 1 # nomor antrian akan selalu bertambah walaupun data sebelumnya telah di hapus(FIFO)

        node_baru = NodePinjam(self.total_antrean, id_p, id_b, nama, nim, prodi, judul, tahun, tgl_p, tgl_k)

        self.storage.append(node_baru) # menambah dan menyimpan node ke dalam list penyimpanan kemudian ditempatkan di bagian selanjutnya (FIFO)
        if not self.head:
            self.head = node_baru
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = node_baru

    def hapus_peminjaman_fifo(self): # Dequeue (Penghapusan/Pengambilan Data), menghapus data yang masuk pertama
        if not self.head:
            print("\n[!!!] Antrean kosong, tidak ada data yang bisa dihapus.")
            return

        data_keluar = self.head # Peek (Melihat data yang ada di paling depan)
        print(f"\n[V] FIFO: Memproses Penyelesain Peminjaman :")
        print(f"    No Antri: {data_keluar.data['no_antri']} | Nama: {data_keluar.data['nama']}")

        self.head = self.head.next
        if self.storage:
            self.storage.pop(0)

        print("[!] Data telah dihapus dari sistem (FIFO).")

    def tampilkan_data(self): # Display (Menampilkan Semua Data)
        if not self.head:
            print(f"\n[!] Belum ada data peminjaman di antrean.")
            print(f"[*] Total buku dipinjam hari ini: {self.total_antrean}")
            return

        print("\n" + "="*95)
        print(f"{'No.Antri':<8} | {'ID':<4} | {'NIM':<8} | {'Nama':<12} | {'Judul Buku':<15}")
        print("-" * 95)

        temp = self.head
        while temp:
            d = temp.data
            print(f"{d['no_antri']:<8} | {d['id_pinjam']:<4} | {d['nim']:<8} | {d['nama']:<12} | {d['judul_buku']:<15}")
            temp = temp.next
        print("="*95)
        print(f"TOTAL TRANSAKSI HARI INI: {self.total_antrean} orang")
```

Menginputkan Sebuah Data

```python
perpus = Linked_List_Perpus()

while True:
    # Perulangan Utama
    print("\nMenu: [1] Tambah Data  [2] Lihat Data  [3] Hapus (FIFO)  [4] Keluar")
    pilihan = input("Pilih menu: ")

    if pilihan == '1': # Enqueue
        print("\n--- Input Data Baru ---")
        id_p = input("ID Pinjam      : ") #id pinjam adalah gabungan dari tgl peminjaman dengan id buku
        id_b = input("ID Buku        : ")
        nama = input("Nama Peminjam  : ")
        nim  = input("NIM            : ")
        prod = input("Prodi          : ")
        jdl  = input("Judul Buku     : ")
        thn  = input("Tahun Terbit   : ")
        tgl_p = input("Tgl Pinjam (YYYY-MM-DD): ")
        tgl_k = input("Tgl Kembali(YYYY-MM-DD): ")

        # Pemanggilan fungsi yang sudah disederhanakan
        perpus.tambah_data(id_p, id_b, nama, nim, prod, jdl, thn, tgl_p, tgl_k)
        print(f"\n[V] Berhasil! Peminjam ini mendapatkan No. Antrian: {perpus.total_antrean}")

    elif pilihan == '2': # Display
        perpus.tampilkan_data()

    elif pilihan == '3': # Dequeue
        perpus.hapus_peminjaman_fifo()

    elif pilihan == '4':
        print("Keluar dari program...")
        break
    else:
        print("Pilihan salah!")
```
