```python
class NodePinjam:
    def __init__(self, id_pinjam, id_buku, nama, nim, prodi, judul, tahun, tgl_pinjam, tgl_kembali):
        self.data = {
            "id_pinjam": id_pinjam,
            "id_buku": id_buku,
            "nama": nama,
            "nim": nim,
            "prodi": prodi,
            "judul_buku": judul,
            "tahun_terbit": tahun,
            "tanggal_pinjam": tgl_pinjam,
            "tanggal_kembali": tgl_kembali
        }
        self.next = None  # Menyimpan indeks atau referensi ke data berikutnya
```

Penyimpanan

```python
class Linked_List_Perpus:
    def __init__(self):
        self.head = None
        self.storage = []

    def tambah_data(self, node_baru):
        self.storage.append(node_baru)
        if not self.head:
            self.head = node_baru
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = node_baru
```
